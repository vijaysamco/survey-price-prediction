# CodeX Project: Data Cleaning

```python
import pandas as pd
import numpy as np

df = pd.read_csv('c:/code/python_jupyter/project_resources/survey_results.csv')
print(df.describe())
```

```python
# --- Step 1: Remove Duplicates ---
cleaned_df = df.drop_duplicates()

# --- Step 2: Outlier Detection and Handling in Age ---
# Unrealistic human age values (> 100) are removed
cleaned_df = cleaned_df[cleaned_df['age'] <= 100]

# --- Step 3: Handling Missing Data ---
# Fill income_levels with 'Not Reported'
cleaned_df['income_levels'] = cleaned_df['income_levels'].fillna("Not Reported")

# Fill missing consume_frequency(weekly) with mode
mode_consume = cleaned_df['consume_frequency(weekly)'].mode()[0]
cleaned_df['consume_frequency(weekly)'] = cleaned_df['consume_frequency(weekly)'].fillna(mode_consume)

# Fill missing purchase_channel with mode
mode_channel = cleaned_df['purchase_channel'].mode()[0]
cleaned_df['purchase_channel'] = cleaned_df['purchase_channel'].fillna(mode_channel)

# --- Step 4: Correcting Spelling Mistakes in Categorical Data ---
# Correct 'zone' inconsistencies
zone_mapping = {
    'urbna': 'Urban',
    'Metor': 'Metro'
}
cleaned_df['zone'] = cleaned_df['zone'].replace(zone_mapping)

# Correct 'current_brand' inconsistencies
brand_mapping = {
    'newcomer': 'Newcomer',
    'Establishd': 'Established'
}
cleaned_df['current_brand'] = cleaned_df['current_brand'].replace(brand_mapping)
cleaned_df.to_csv('c:/code/python_jupyter/project_resources/cleaned_survey_results.csv', index=False)
print("Data cleaning complete.")
```
